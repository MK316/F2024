## **[4] Python introduction (Audio)**

```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text')
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

def etts(txt, filename):
    gtts_object = gTTS(text=txt, lang="en", tld="us", slow=False)
    gtts_object.save(filename)
def ktts(txt, filename):
    gtts_object = gTTS(text=txt, lang="ko", slow=False)
    gtts_object.save(filename)
```

```python
#@markdown Introduction to Python: 7 Audio files
from IPython.display import Audio, display

texts = [
    "Welcome! I'm Gee Tee, your digital assistant for 'Digital Literacy and English Education'. Today, we explore Python programming, a crucial skill for enhancing digital literacy in your education career.",
    "Python's simplicity makes it ideal for beginners. As future educators, understanding coding is key to engaging digital-native students effectively. We'll start with the basics, aiming to make your teaching dynamic. Letâ€™s embrace this journey together, growing our skills and understanding of the digital landscape. Welcome aboard!"
]

# Loop through each text, generate audio, and save it
for i, text in enumerate(texts):
    filename = f"E-audio_{i+1}.mp3"
    etts(text, filename)
    print(filename)
    print("*"*30)
    display(Audio(filename))
```

```python
#@markdown Sound files (Elevenlabs)
# First, install necessary libraries if not already available
# !pip install librosa

import librosa
import requests
from io import BytesIO
from IPython.display import Audio

file = "2" #@param = ["1","2"]
fn = int(file)-1

# URL of the MP3 file on GitHub
urls = ['https://raw.githubusercontent.com/MK316/F2024/main/Multimedia/data/eleven-sample1.mp3',
             'https://raw.githubusercontent.com/MK316/F2024/main/Multimedia/data/eleven-sample2.mp3']

audio_url = urls[fn]
# Download the audio file from the GitHub link
response = requests.get(audio_url)
audio_bytes = BytesIO(response.content)

# Load the audio file with librosa
audio, sr = librosa.load(audio_bytes, sr=None)  # `sr=None` loads the file with its original sampling rate

# Display an audio player for the loaded audio
Audio(data=audio, rate=sr)
```
